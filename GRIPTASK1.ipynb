**Task1 - Prediction Using Supervised ML**

Simple Linear Regression to predict the scores of a student based on the number of study hours 

### **Importing the libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### **Importing the dataset**

```python
dataset = pd.read_csv("http://bit.ly/w-data")
X = dataset.iloc[:, :-1].values #take values from all the rows and columns excpet last column i.e independent variable
y = dataset.iloc[:, -1].values  #take values of the dependent variable only i.e last column only

dataset.head(10)
```

```python
dataset.isnull == True      #checking for missing values in the data
```

Thus there are no missing values.

### **Exploring the details of the dataset**

```python
print(dataset.describe())       #details of the dataset
```

### **Visualizing the data**

```python
sns.scatterplot(x = 'Scores', y = 'Hours', data = dataset)
```

From the above graph we can observe that there seems to be linear relationship between the number of study hours and the scores.

### **Splitting the dataset into the Training set and Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### **Training the Simple Linear Regression model on the Training set**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train) 
```

### **Visualising the Training set results**

```python
plt.scatter(X_train, y_train, color = 'red' )
plt.plot(X_train, regressor.predict(X_train), color = 'blue' )
plt.title('Scores Vs No. of study hours(Training Set)')
plt.xlabel('Scores')
plt.ylabel('No. of study hours') 
```

### **Visualising the Test set results**

```python
plt.scatter(X_test, y_test, color = 'red' )                      #scattered the test set values
plt.plot(X_train, regressor.predict(X_train), color = 'blue' )   #plotted the training set values
plt.title('Percentage Vs No. of Training Hours (Test Set)')
plt.xlabel('Percentage')
plt.ylabel('No. of study hours') 
```

### **Making Predictions**

```python
print(X_test) # Testing data - In Hours
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
print(df) 
```

```python
hours = ([[9.25]])                  #2D array
own_pred = regressor.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

### **Evaluating the model - how accurate is our model?**

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred))
```

```python
from sklearn import metrics
print("Accuracy of the model: ")
metrics.r2_score(y_test, y_pred)
```
